```python
import shutil
from os import walk

def hasNumber(inputString):
    return any(char.isdigit() for char in inputString)

srcPath = 'recHW20796\\'
targetPath = 'recHW20796SubEntry\\'
f = []
for (dirpath, dirnames, filenames) in walk(srcPath):
    f.extend(filenames)
    break

for word in f:
    if hasNumber(word):
        print word
        shutil.move(srcPath + word, targetPath + word)
    
```
