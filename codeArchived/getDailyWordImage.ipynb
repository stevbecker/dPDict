```python
import urllib
import urllib2
from bs4 import BeautifulSoup
import codecs

response = urllib2.urlopen('http://www.learnersdictionary.com/word-of-the-day/archives')
html = response.read()
#print html
#<tr class= "wod_archives_list_entry">
#               <td class = "wod_archives_list_entry_td1">
#                    November 24th                </td>
#                <td class = "wod_archives_list_entry_td2">
#<a class="wod_archives_list_entry_td2_link1" href="/word-of-the-day/2015/11/24">free-for-all</a>
#                </td>
#            </tr>''' 

#soup = BeautifulSoup(html, 'html.parser')
soup = BeautifulSoup(html)

#print soup.title
#print(soup.prettify())

linkword = {}
lws = []
cnt = 0

for link in soup.find_all('a'):    
    if link.get('href').find('word-of-the-day/') != -1:
        cnt += 1
        
        linkword['link'] = link.get('href')
        linkword['word'] = link.string
        
        lws.append(linkword.copy())
        #print cnt, ': ', linkword, '...', lws, '\n\n'
print cnt

fimg = codecs.open("E:\\1Now\\dPDict\\processDict\\anki\\img\\dailyimg.txt", 'w', 'utf-8')
cnt = 0
for lw in lws:
    cnt += 1
    #if cnt > 1: break # test one link
        
    print cnt, '...', lw['link'], '...', lw['word']
    resp = urllib2.urlopen('http://www.learnersdictionary.com' + lw['link'])
    soup = BeautifulSoup(resp.read())
    for imglist in soup.find_all('img'):
        if imglist.get('src').find('word_of_the_day/') != -1:
            imgurl = 'http://www.learnersdictionary.com' + imglist.get('src')
            print imgurl
            print imglist.get('alt')
            #urllib.urlretrieve(imgurl, "E:\\1Now\\dPDict\\processDict\\anki\\img\\" + lw['word'] + '.jpg')
            fimg.write(lw['word'] + '.jpg;' + imglist.get('alt') + '\n')

fimg.close()

#<!--Image and title-->
#                   <div class="wod_img_act"><img src = "/media/ld/word_of_the_day/images/2682/large.jpg" alt = "A&#x20;snowball&#x20;fight&#x20;that&#x20;turned&#x20;into&#x20;a&#x20;free-for-all"></div>
#           <div class="wod_img_tit">A snowball fight that turned into a free-for-all</div>'''        
```
