```python
#get the list of basic words and their inflections from MongoDB.
#Basic words are 3776; plus their inflections, there're 8751

import pymongo
from pymongo import MongoClient

import shutil


fbasicwords = open('subDict\\basicwords.txt', 'w')


#connect to MongoDB
client = MongoClient('localhost', 27017)
db = client.dpdb
dpdict = db.dpdict


cnt = 1
uniBW = set()
basicwords = dpdict.find({"bw":1}).sort("hw")
for bw in basicwords:    
    hw = bw.get("hw") #get is a method of dictionary
    print hw
    uniBW.add(hw)
    if bw.get("infl"):
        infls = set(bw.get("infl"))
        for infl in infls:
            #if infl != hw: print infl
            uniBW.add(infl)
        
    cnt += 1
    #if cnt > 100: break
print cnt

for bw in sorted(uniBW):
    #print bw
    fbasicwords.write(bw+"\n")



    
    
    
```
